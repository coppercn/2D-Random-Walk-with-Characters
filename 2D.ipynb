```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(steps):
    x = np.zeros(steps)
    y = np.zeros(steps)
    
    for i in range(1, steps):
        direction = np.random.randint(1, 5)
        if direction == 1:
            x[i] = x[i-1] + 1
            y[i] = y[i-1]
        elif direction == 2:
            x[i] = x[i-1] - 1
            y[i] = y[i-1]
        elif direction == 3:
            x[i] = x[i-1]
            y[i] = y[i-1] + 1
        else:
            x[i] = x[i-1]
            y[i] = y[i-1] - 1
    
    return x, y
```

```python
def plot_random_walks(steps):
    plt.figure(figsize=(8,6))

    x, y = random_walk(steps)
    plt.plot(x, y, "b--", x[0], y[0], "ro", x[-1], y[-1], "r*")
    
    plt.title(f"2D Random Walks in {steps} Steps")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()
```

```python
plot_random_walks(1000)
```

```python
plot_random_walks(100000)
```
