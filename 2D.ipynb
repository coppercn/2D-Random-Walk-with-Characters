```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(steps):
    x = np.zeros(steps)
    y = np.zeros(steps)
    
    for i in range(1, steps):
        direction = np.random.randint(0, 4)
        if direction == 0:   # Move right
            x[i] = x[i-1] + 1
            y[i] = y[i-1]
        elif direction == 1: # Move left
            x[i] = x[i-1] - 1
            y[i] = y[i-1]
        elif direction == 2: # Move up
            x[i] = x[i-1]
            y[i] = y[i-1] + 1
        else:                # Move down
            x[i] = x[i-1]
            y[i] = y[i-1] - 1
    
    return x, y
```

```python
def plot_random_walks(steps):
    x, y = random_walk(steps)
    
    plt.plot(x, y, "b--", label='Path', zorder=1)
    plt.scatter(x[0], y[0], c="red", marker='o', label='Starting point', zorder=2)
    plt.scatter(x[-1], y[-1], c="red", marker='*', label='End point', zorder=2)
    
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title(f"2D Random Walks in {steps} Steps")
    plt.legend()
    plt.show()
```

```python
plot_random_walks(1000)
```

```python
plot_random_walks(100000)
```
